```python
def display_results(model):
    print("\u0332".join("Optimal Lineup:\n"))
    score = str(model.objective)
    for v in model.variables():
        score = score.replace(v.name, str(v.varValue))
        # Print Players by positon
        if v.varValue != 0:
            print(v.name)
    print("\nTotal Points: {:.2f}".format(eval(score)))
```
