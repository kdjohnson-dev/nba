```python
from datetime import datetime
import pytz
import time
import requests
import bs4
```

```python
base_url = 'https://www.espn.com/nba/schedule'
res = requests.get(base_url)
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
day = 0  # track week day
group = 0  # track group number
groups = soup.select(".responsive-table-wrap")
dates = soup.select("#sched-container")[0].select('h2')
print_flag = True
group_flag = True if len(groups) == 8 else False
pst = pytz.timezone('US/Pacific')

while day < 7: 
    # Display day of the week
    if print_flag: print("\u0332".join("{}\n ".format(dates[day].text)))
    
    # Get all games in given group
    games = groups[group].select('tr')
    
    # Get and display daily game information
    for index in range(1, len(games)):
        # Get team matchups
        away = games[index].select('a')[1].select('abbr')[0]['title']
        home = games[index].select('a')[3].select('abbr')[0]['title']
        
        # Get game status (game times or scores) 
        if 'result' in str(games):
            status = games[index].select('a')[4].text
        elif 'live' in str(games[index]):
            status = games[index].select('a')[4].text
        else:
            date_time = games[index].select('td')[2].attrs['data-date']
            date_time = datetime.fromisoformat(date_time[:-1])
            date_time = pytz.timezone('UTC') \
                            .localize(date_time) \
                            .astimezone(pst) \
                            .strftime('%H:%M') 
            
            status = time.strptime(date_time, '%H:%M')
            status = time.strftime("%I:%M: %p", status)
            if status[0] == '0': status = status[1:]
            
        # Display matchups and game times/scores 
        print("{: <30}\tat\t{: <30} @\t{}".format(away,home,status))
    
    # Check for finished games on current day
    if group_flag:
        if ('result' not in str(groups[group])) and ('result' in str(groups[group+1])):
            print_flag = False
            group_flag = False
            group += 1
            continue
    
    group += 1
    day += 1 
    print_flag = True
    print("\n")
```
